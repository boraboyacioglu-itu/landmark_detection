```python
from matplotlib import pyplot as plt
```

```python
name = 'fatmagul'
with open(f'landmarks/{name}.txt', 'r') as f:
    landmarks = f.readlines()
```

```python
df = []
for landmark in landmarks:
    points = landmark[2:-3].split('), (')
    lm = []
    for point in points:
        x, y = point.split(', ')
        lm.append([int(x), int(y)])
    df.append(lm)
```

```python
for i, lm in enumerate(df):
    print(f'Iteration: {i}')
    plt.figure()
    for x in lm:
        plt.scatter(x[0], x[1], c='b', s=1)
    plt.show()
    plt.close()
```
